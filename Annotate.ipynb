```python
import pandas as pd
import glob
import os
```

```python
final_df = []

root = 'UECFOOD256/'
write_txt =  '/datasets/work-thanawit/UECFOOD256/'
#Load Notation File 
category_name = pd.read_table(root+'category.txt')

#Read all file
for i in range(1,257):
    current_frame = []
    current_image_path = str(i)
    annotate_location = glob.glob(os.path.join(root+current_image_path , "*.txt"))
    current_file = pd.read_csv(annotate_location[0], sep=" ", header=None, names = ['img', 'x1', 'y1','x2','y2'])
    current_file = current_file.iloc[1:]
    #print(first_file)
    #current_annotate = pd.read_table(annotate_location[0])
    #current_frame.append(current_annotate.img)
    for index,row in current_file.iterrows():
        idx = []
        #img_name, x1, y1,x2,y2
        idx = write_txt+str(i)+"/"+str(str(row[0])+".jpg"+" "+str(row[1])+","+str(row[2])+","+str(row[3])+","+str(row[4])+","+str(i-1))
        final_df.append(idx)
    

    

with open('orange_final_annotate.txt', 'w') as f:
    for item in final_df:
        f.write("%s\n" % item)
```
